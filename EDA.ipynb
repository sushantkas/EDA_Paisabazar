### Categorical Variables Visualization

```python
# Checking countwise distribution for Categroical columns skipping Columns having more than 20 unique values
for i in df[objtype]:
    if df[i].unique().shape[0] <= 20:
        plt.title(f"Count Plot for {i} Column")
        sns.countplot(df[i], legend=True)
        plt.show()
    else:
        print(f"Skipping {i} Column")
```

Categorical Inference : <br>
Occupation columns has almost same number of values  <br>

Payment of min columns has almost same number of values <br>
Payment_Behaviour columns has almost same number of values <br>
Credit_Score columns has 3 values Standard, Good, Poor standard has highest count <br>

```python
# create figure with fixed size
plt.figure(figsize=(15, 25))

j = 1
for i in inttype:
    if len(df[i].unique())<45:
        try:
            plt.subplot(4, 2, j)   # 3 rows, 2 cols, position j
            sns.countplot(data=df, x=i)  # kde=True for smooth curve (optional)
            plt.title(f"Countplot of {i}")
            j += 1
        except:
            print(f"Skipping {i} Column")
    else:
        print(f"Skipping {i} Column Because of Unique value more than 45")

plt.tight_layout()  # adjust spacing
plt.show()
```
